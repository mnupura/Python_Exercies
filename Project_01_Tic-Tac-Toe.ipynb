```python
board=[1,2,3,4,5,6,7,8,9]
player1=''
player2=''
player_flag=True

#Display the Tic-Tac-Toe board
def board_display():
    print(str(board[0])+'|'+str(board[1])+'|'+str(board[2]))
    print('-----')
    print(str(board[3])+'|'+str(board[4])+'|'+str(board[5]))
    print('-----')
    print(str(board[6])+'|'+str(board[7])+'|'+str(board[8]))

#Players to select options(X,O)
def player_select():
    global player1
    global player2
    check_player_input=True
    while check_player_input: 
        print('Select option (X,O):')
        player1=input('Player 1: ')
        if player1=='X':
            player2='O'
            print('Player 2: {}'.format(player2))
            check_player_input=False
        elif player1=='O':
            player2='X'
            print('Player 2: {}'.format(player2))
            check_player_input=False
        else:    
            print('Please enter valid input: (X,O)')
        

#Enter options on the positions
def enter_option():
    global player_flag
    check_position=True
    while check_position:
        try:
            if player_flag==True:
                position=int(input('Enter position (Player 1): '))
                check_position=position_check(position,player1)       
            else:
                position=int(input('Enter position (Player 2): '))     
                check_position=position_check(position,player2)
        except:
                print('Incorrect Input. Re-enter position.')
    board_display()

#Checks for - the position on the board is occupied or not, incorrect position entered. Also,update position if all is ok.
def position_check(position,player):
    global player_flag
    if position>=10 and position<0:
        print('Incorrect position. It should be within 1 and 9. Re-enter position.')
        return True
    elif board[position-1] in ['X','O']:
        print('Position already occupied. Re-enter position.')
        return True
    else:
        board[position-1]=player
        player_flag=not player_flag       
        return False        

    
#Update options on the board
def board_update():
    length=9
    winner_found=False
    result=''
    while length>0:
        enter_option()
        if game_winner(player1):
            result='Game over! Winner is Player 1!'
            print(result)
            length=0
            winner_found=True
        elif game_winner(player2):
            result='Game over! Winner is Player 2!'
            print(result)
            length=0
            winner_found=True
        else:
            length=length-1
    if winner_found==False:
        result='Game draw!'
        print(result)
        
    result_file(result)    
    play_again()

#Returns result if game is over and there is a winner
def game_winner(player):
    return ((board[0]==board[1]==board[2]==player) or (board[0]==board[3]==board[6]==player) or 
            (board[0]==board[4]==board[8]==player) or (board[3]==board[4]==board[5]==player) or 
            (board[6]==board[7]==board[8]==player) or (board[1]==board[4]==board[7]==player) or 
            (board[2]==board[5]==board[8]==player) or (board[6]==board[4]==board[2]==player))
            
#Ask users if they wish to continue the game or stop.
def play_again():
    play_again_input=True
    while play_again_input: 
        replay=input('Would you like to play again? (Yes/No): ')
        if replay=='Yes':
            global board
            board=[1,2,3,4,5,6,7,8,9]
            global player1
            player1=''
            global player2
            player2=''
            global player_flag
            player_flag=True
            play_game_start()
            play_again_input=False
        elif replay=='No':
            print('Game stopped as per you request!')
            play_again_input=False
        else:
            print('Unable to understand the input. Please re-enter.')

#Save the result of the game in a file
def result_file(result):
    game_result_file=open('Project_01_Result_File.txt',mode='a')
    game_result_file.write('\n'+result)
    game_result_file.close()
    
    game_result_file=open('Project_01_Result_File.txt',mode='r')
    print(game_result_file.read())
    game_result_file.close()
        
#Start game        
def play_game_start():
    board_display()
    player_select()
    board_update()

play_game_start()
```
