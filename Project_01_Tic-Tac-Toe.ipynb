```python
board=[1,2,3,4,5,6,7,8,9]
player1=''
player2=''

#Display the Tic-Tac-Toe board
def board_display():
    print(str(board[0])+'|'+str(board[1])+'|'+str(board[2]))
    print('-----')
    print(str(board[3])+'|'+str(board[4])+'|'+str(board[5]))
    print('-----')
    print(str(board[6])+'|'+str(board[7])+'|'+str(board[8]))

#Players to select options(X,O)
def player_select():
    print('Select option (X,O):')
    player1=input('Player 1: ')
    if player1=='X':
        player2='O'
    else:
        player2='X'
    print('Player 2: {}'.format(player2))

#Enter options on the positions
def enter_option():
    position=input('Enter position: ')
    option=input('Enter option (X,O): ')
    board[int(position)-1]=option
    board_display()

#Update options on the board
def board_update():
    length=9
    while length>0:
        enter_option()
        if (board[0]==board[1]==board[2]=='X') or (board[0]==board[3]==board[6]=='X') or (board[0]==board[4]==board[8]=='X') or (board[3]==board[4]==board[5]=='X') or (board[6]==board[7]==board[8]=='X') or (board[1]==board[4]==board[7]=='X') or (board[2]==board[5]==board[8]=='X') or (board[6]==board[4]==board[2]=='X'):
            if player1=='X':
                print('Game over! Winner is Player 1!')
            else:
                print('Game over! Winner is Player 2!')
            length=0
        elif (board[0]==board[1]==board[2]=='O') or (board[0]==board[3]==board[6]=='O') or (board[0]==board[4]==board[8]=='O') or (board[3]==board[4]==board[5]=='O') or (board[6]==board[7]==board[8]=='O') or (board[1]==board[4]==board[7]=='O') or (board[2]==board[5]==board[8]=='O') or (board[6]==board[4]==board[2]=='O'):
            if player1=='O':
                print('Game over! Winner is Player 1!')
            else:
                print('Game over! Winner is Player 2!')
            length=0
        length=length-1

board_display()
player_select()
board_update()
```
