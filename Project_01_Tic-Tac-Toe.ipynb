```python
board=[1,2,3,4,5,6,7,8,9]
player1=''
player2=''
player_flag=True

#Display the Tic-Tac-Toe board
def board_display():
    print(str(board[0])+'|'+str(board[1])+'|'+str(board[2]))
    print('-----')
    print(str(board[3])+'|'+str(board[4])+'|'+str(board[5]))
    print('-----')
    print(str(board[6])+'|'+str(board[7])+'|'+str(board[8]))

#Players to select options(X,O)
def player_select():
    global player1
    global player2
    print('Select option (X,O):')
    player1=input('Player 1: ')
    if player1=='X':
        player2='O'
    else:
        player2='X'
    print('Player 2: {}'.format(player2))

#Enter options on the positions
def enter_option():
    global player_flag
#    try:
    if player_flag==True:
        position=int(input('Enter position (Player 1): '))
        if position<10 and position>0:
            board[position-1]=player1
            player_flag=False
        else:
            print('Incorrect Input. Re-enter position.')
    else:
        position=int(input('Enter position (Player 2): '))     
        if position<10 and position>0:
            board[position-1]=player2
            player_flag=True
        else:
            print('Incorrect Input. Re-enter position.')
#    except:
#        print('Incorrect Input. Re-enter position.')
    board_display()

#Update options on the board
def board_update():
    length=9
    winner_found=False
    while length>0:
        enter_option()
        if game_winner(player1):
            print('Game over! Winner is Player 1!')
            length=0
            winner_found=True
        elif game_winner(player2):
            print('Game over! Winner is Player 2!')
            length=0
            winner_found=True
        else:
            length=length-1

    if winner_found==False:
        print('Game draw!')
    play_again()


def game_winner(player):
    return ((board[0]==board[1]==board[2]==player) or (board[0]==board[3]==board[6]==player) or 
            (board[0]==board[4]==board[8]==player) or (board[3]==board[4]==board[5]==player) or 
            (board[6]==board[7]==board[8]==player) or (board[1]==board[4]==board[7]==player) or 
            (board[2]==board[5]==board[8]==player) or (board[6]==board[4]==board[2]==player))
            
def play_again():
    replay=input('Would you like to play again? (Yes/No): ')
    if replay=='Yes':
        global board
        board=[1,2,3,4,5,6,7,8,9]
        global player1
        player1=''
        global player2
        player2=''
        global player_flag
        player_flag=True
        play_game_start()
    else:
        print('Game stopped as per you request!')
        
def play_game_start():
    board_display()
    player_select()
    board_update()

play_game_start()
```
